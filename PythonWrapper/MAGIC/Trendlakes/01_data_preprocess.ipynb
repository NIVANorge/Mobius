```python
import pandas as pd
```

```python
timeseries_df = pd.read_csv('raw_data/water_soil_chem_hydro_depos_df.csv')
list(timeseries_df.keys())
```

```python
timeseries_df.keys()
```

```python
# Convert the 'Date' column to datetime format
timeseries_df['Date'] = pd.to_datetime(timeseries_df['Date'])

# Set the 'Date' column as the index
timeseries_df.set_index('Date', inplace=True)

# Drop the rows for the years 1880 and 1881
timeseries_df = timeseries_df.loc[~timeseries_df.index.year.isin([1880, 1881])]

# Reset the index
timeseries_df.reset_index(inplace=True)
timeseries_df
```

```python
# timeseries_df.drop('Unnamed: 0', axis=1, inplace=True)
timeseries_df.info()
```

```python
timeseries_df.to_csv("raw_data/water_soil_chem_hydro_depos_corrected.csv", index=False)
```
